# Functions

```python
def calculate_area(length, breadth):
    return length * breadth
print(calculate_area(5,4))
```

```python
# function to check prime or not

def is_prime(num):
    if num <= 1:
        return False
    for i in range(2,num):
        if num % i == 0:
            return False
    return True
def main():
    num = int(input("enter no"))
    if is_prime(num):
        print("prime")
    else:
        print("not prime")
main()    
```
