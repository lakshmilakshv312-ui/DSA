# Functions

```python
def calculate_area(length, breadth):
    return length * breadth
print(calculate_area(5,4))
```

```python
# function to check prime or not

def is_prime(num):
    if num <= 1:
        return False
    for i in range(2,num):
        if num % i == 0:
            return False
    return True
def main():
    num = int(input("enter no"))
    if is_prime(num):
        print("prime")
    else:
        print("not prime")
main()    
```

# Arrays

* An array is a collection of elements stored in a contigenous memory location.

```python
# Example
arr = [10, 20, 30, 40, 50]
print(arr)

# Access Elements
print(arr[0])   # first element → 10
print(arr[2])   # third element → 30

# Modify Elements
arr[1] = 25
print(arr)

# Add Elements
arr.append(60)      # add at end
arr.insert(1, 15)   # add at index

# Remove Elements
arr.remove(30)   # remove value
arr.pop()        # remove last

#Loop Through Array
for i in arr:
    print(i)

# User Input Array
arr = list(map(int, input("Enter numbers: ").split()))
print(arr)
```

```python
# Taking user input
arr = list(map(int, input("Enter numbers: ").split()))

# Display array
print("Array:", arr)

# Access elements
print("First element:", arr[0])
print("Last element:", arr[-1])

# Modify element
if len(arr) > 1:
    arr[1] = 999
print("After modifying index 1:", arr)

# Add elements
arr.append(100)         # add at end
arr.insert(1, 50)       # add at index
print("After adding elements:", arr)

# Remove elements
if 50 in arr:
    arr.remove(50)      # remove specific value
if len(arr) > 0:
    arr.pop()           # remove last element
print("After removing elements:", arr)

# Searching element
key = int(input("Enter element to search: "))
if key in arr:
    print("Element found")
else:
    print("Element not found")

# Count frequency
print("Frequency of", key, ":", arr.count(key))

# Find max and min
print("Maximum:", max(arr))
print("Minimum:", min(arr))

# Sorting
arr.sort()
print("Sorted array:", arr)

# Reverse
arr.reverse()
print("Reversed array:", arr)

# Looping
print("Array elements:")
for i in arr:
    print(i, end=" ")
```
