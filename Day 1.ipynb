```python
# FOR LOOP
def for_loop_demo(a):
    print("For loop output:")
    for i in a:
        print(i, end=" ")
    print()

# WHILE LOOP
def while_loop_demo(a):
    print("While loop output:")
    i = 0
    while i < len(a):
        print(a[i], end=" ")
        i += 1
    print()

# SUM using loop
def sum_elements(a):
    s = 0
    for i in a:
        s += i
    return s

# COUNT using loop
def count_elements(a):
    count = 0
    for i in a:
        count += 1
    return count

# MAX using loop
def max_element(a):
    m = a[0]
    for i in a:
        if i > m:
            m = i
    return m

# MIN using loop
def min_element(a):
    m = a[0]
    for i in a:
        if i < m:
            m = i
    return m

# MAIN
arr = list(map(int, input("Enter numbers: ").split()))

for_loop_demo(arr)
while_loop_demo(arr)

print("Sum:", sum_elements(arr))
print("Count:", count_elements(arr))
print("Max:", max_element(arr))
print("Min:", min_element(arr))
```
