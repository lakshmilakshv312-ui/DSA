```python
# counting problems
# count even no from 1 to N
def count_even(n):# parameter
    count = 0
    for i in range(1,n+1):
        if i % 2 == 0:
            count += 1
    return count
def main():
    n = int(input("enter the no:"))
    print("even count:",count_even(n))
main()
```

```python
# count digits in numbner
# input: 12345
# output : 5

def count_digits(num):
    count = 0
    while num > 0:
        num = num // 10
        count += 1
    return count
def main():
    num = int(input("enter number:"))
    print("digits count:", count_digits(num))
main()
```
