```python
# counting problems
# count even no from 1 to N
def count_even(n):# parameter
    count = 0
    for i in range(1,n+1):
        if i % 2 == 0:
            count += 1
    return count
def main():
    n = int(input("enter the no:"))
    print("even count:",count_even(n))
main()
```

```python
# count digits in numbner
# input: 12345
# output : 5

def count_digits(num):
    count = 0
    while num > 0:
        num = num // 10
        count += 1
    return count
def main():
    num = int(input("enter number:"))
    print("digits count:", count_digits(num))
main()
```

```python
# counting problems
# Function to count total elements
def total_count(a):
    return len(a)

# Function to count even and odd
def even_odd_count(a):
    even = odd = 0
    for i in a:
        if i % 2 == 0:
            even += 1
        else:
            odd += 1
    return even, odd

# Function to count positive, negative, zero
def pos_neg_zero(a):
    pos = neg = zero = 0
    for i in a:
        if i > 0:
            pos += 1
        elif i < 0:
            neg += 1
        else:
            zero += 1
    return pos, neg, zero

# Function to count frequency
def frequency(a, key):
    return a.count(key)

# Function to count duplicates
def duplicate_count(a):
    visited = []
    dup = 0
    for i in a:
        if i not in visited:
            if a.count(i) > 1:
                dup += 1
            visited.append(i)
    return dup

# Function to count unique elements
def unique_count(a):
    count = 0
    for i in a:
        if a.count(i) == 1:
            count += 1
    return count

# MAIN
arr = list(map(int, input("Enter numbers: ").split()))

print("Total:", total_count(arr))

even, odd = even_odd_count(arr)
print("Even:", even, "Odd:", odd)

pos, neg, zero = pos_neg_zero(arr)
print("Positive:", pos, "Negative:", neg, "Zero:", zero)

key = int(input("Enter element: "))
print("Frequency:", frequency(arr, key))

print("Duplicates:", duplicate_count(arr))
print("Unique:", unique_count(arr))
```
