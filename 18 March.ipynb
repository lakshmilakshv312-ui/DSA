```python
# prefix sum problems
def prefix_sum(arr):
    arr = [1,2,3,4,5]
    prefix = [0] * len(arr)
    prefix[0] = arr[0]

    for i in range(1, len(arr)):
        prefix[i] = prefix[i-1] + arr[i]

    return prefix
arr = [1,2,3,4,5]
print("Prefix sum:", prefix_sum(arr))
```

```python
# palindrome
def is_palindrome(string):
    left = 0
    right = len(string) - 1

    while left < right:
        if string[left]!= string[right]:
            return False
        left += 1
        right -= 1
    return True
string = "racecar"
print("palindrome:", is_palindrome(string))
        
```
